```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

# Drought Data Analyzation

```python
#United States Dataset
#Load datasets
droughts = open('data/drought_monitor_data/dm_export.csv')
drought_data_US = pd.read_csv(droughts, parse_dates= ['MapDate'])
drought_data_US= drought_data_US[::-1]
drought_data_US.index = drought_data_US["MapDate"]
drought_data_US = drought_data_US.drop('MapDate', axis = 1)
#Build a linear model to function as a mean line. 
drought_data_arr = np.array(drought_data_US['D4'].values)
x = np.array(np.arange(0,1201))
x = x.reshape(-1,1)
print(x.shape, drought_data_arr.shape)
lr = LinearRegression().fit(x,drought_data_arr)
#Plot Everything
fig = plt.figure(figsize = (50,40))
plt.plot(drought_data_US.index,drought_data_US['D4'])  
plt.plot(drought_data_US.index,lr.predict(x))
plt.xticks(drought_data_US.index[::14], rotation = 60, fontsize = 24)
plt.title('D4 Drought percentages as a function of year between 2000-2023 in the United States', fontsize = 40)
plt.hlines
plt.yticks(fontsize = 25)
plt.xlabel('Dates, in Datetime Format', fontsize = 30)
plt.ylabel('D4 drought percentage, in percent', fontsize = 30)
plt.legend(['Drought trend','Mean Rate of change of D4 prevelance'], fontsize = 40)
plt.show()
drought_data_US['D4'].values
```

```python
#Nevada Dataset
#Load datasets
droughts2 = open('data/drought_monitor_data/dm_export_Nevada.csv')
drought_data_NV = pd.read_csv(droughts2, parse_dates= ['MapDate'])
drought_data_NV= drought_data_NV[::-1]
drought_data_NV.index = drought_data_NV["MapDate"]
drought_data_NV = drought_data_NV.drop('MapDate', axis = 1)
#Build a linear model
drought_data_arr = np.array(drought_data_NV['D2'].values)
x = np.array(np.arange(0,1208))
x = x.reshape(-1,1)
print(x.shape, drought_data_arr.shape)
lr = LinearRegression().fit(x,drought_data_arr)
#Plot Everything
fig = plt.figure(figsize = (50,40))
plt.plot(drought_data_NV.index,drought_data_NV['D2'])  
plt.plot(drought_data_NV.index,lr.predict(x))
plt.xticks(drought_data_NV.index[::14], rotation = 60, fontsize = 24)
plt.title('D2 Drought percentages as a function of year between 2000-2023 in Nevada, USA', fontsize = 40)
plt.yticks(fontsize = 25)
plt.xlabel('Dates, in Datetime Format', fontsize = 30)
plt.ylabel('D2 drought percentage, in percent', fontsize = 30)
plt.legend(['Drought trend','Rate of change of D2 prevelance'], fontsize = 40)
plt.show()
```

```python
#Colorado Dataset
#Load datasets
droughts2 = open('data/drought_monitor_data/dm_export_Colorado.csv')
drought_data_CO = pd.read_csv(droughts2, parse_dates= ['MapDate'])
drought_data_CO= drought_data_CO[::-1]
drought_data_CO.index = drought_data_CO["MapDate"]
drought_data_CO = drought_data_CO.drop('MapDate', axis = 1)
#Build a linear model
drought_data_arr = np.array(drought_data_CO['None'].values)
x = np.array(np.arange(0,1201))
x = x.reshape(-1,1)
print(x.shape, drought_data_arr.shape)
lr = LinearRegression().fit(x,drought_data_arr)
#Plot Everything
fig = plt.figure(figsize = (50,40))
plt.plot(drought_data_CO.index,drought_data_CO['None']) 
plt.plot(drought_data_CO.index, lr.predict(x)) 
plt.xticks(drought_data_CO.index[::14], rotation = 60, fontsize = 24)
plt.title('No Drought percentages as a function of year between 2000-2023 in Colorado, USA', fontsize = 40)
plt.yticks(fontsize = 25)
plt.xlabel('Dates, in Datetime Format', fontsize = 30)
plt.ylabel('No drought percentage, in percent', fontsize = 30)
plt.legend(['Drought trend','Rate of change of no drought prevelance'], fontsize = 40)
plt.show()
```

```python
#Statistics USA
print('D4 drought averages in the US during the 2000s,2010s, and 2020s')
drought_data_US['year'] = drought_data_US.index
drought_data_US_2000s = drought_data_US[drought_data_US['year'] < pd.to_datetime('2009-12-30')]
print(drought_data_US_2000s['D4'].mean())

drought_data_US_2010s1 = drought_data_US[drought_data_US['year'] < pd.to_datetime('2019-12-30')]
drought_data_US_2010s2 = drought_data_US_2010s1[drought_data_US_2010s1['year'] > pd.to_datetime('2009-12-30')]
print(drought_data_US_2010s2['D4'].mean())

drought_data_US_2020s = drought_data_US[drought_data_US['year'] > pd.to_datetime('2019-12-30')]
print(drought_data_US_2020s['D4'].mean(),'\n')

#Statistics Co
print('D4 drought averages in Colorado during the 2000s,2010s, and 2020s')
drought_data_CO['year'] = drought_data_CO.index
drought_data_Co_2000s = drought_data_CO[drought_data_CO['year'] < pd.to_datetime('2009-12-30')]
print(drought_data_Co_2000s['D4'].mean())

drought_data_CO_2010s1 = drought_data_CO[drought_data_CO['year']< pd.to_datetime('2019-12-30')]
drought_data_CO_2010s2 = drought_data_CO[drought_data_CO['year']> pd.to_datetime('2009-12-30')]
print(drought_data_CO_2010s2['D4'].mean())

drought_data_CO_2020s = drought_data_CO[drought_data_CO['year'] > pd.to_datetime('2019-12-30')]
print(drought_data_CO_2020s['D4'].mean(),'\n')

# #Statistics NV
print('D4 drought averages in Nevada during the 2000s,2010s, and 2020s')
drought_data_NV['year'] = drought_data_NV.index
drought_data_NV_2000s = drought_data_NV[drought_data_NV['year'] < pd.to_datetime('2009-12-30')]
print(drought_data_NV_2000s['D4'].mean())

drought_data_NV_2010s1 = drought_data_NV[drought_data_NV['year'] < pd.to_datetime('2019-12-30')]
drought_data_NV_2010s2 = drought_data_NV[drought_data_NV['year'] > pd.to_datetime('2009-12-30')]
print(drought_data_NV_2010s2['D4'].mean())

drought_data_NV_2020s = drought_data_NV[drought_data_NV['year'] > pd.to_datetime('2019-12-30')]
print(drought_data_NV_2020s['D4'].mean(),'\n')
```

```python
fig_f, axes = plt.subplots(2,3, figsize = (25,25))
[ax1,ax2,ax3,ax4,ax5,ax6] = axes.flatten()

#Colorado d4 drought plot by year
drought_data_CO['year'] = drought_data_CO.index.year
ax1.plot(drought_data_CO.groupby(['year']).mean()['D4'])
ax1.set_title('Yearly D4 drought averages in Colorado \n', fontsize = 15)
ax1.set_xlabel('Years of the 21st Century', fontsize = 15)
ax1.set_ylabel('D4 drought average, in %', fontsize = 15)

drought_data_CO['month'] = drought_data_CO.index.month
ax4.plot(drought_data_CO.groupby(['month']).mean()['D4'])
ax4.set_title('Monthly D4 drought averages in Colorado \n', fontsize = 15)
ax4.set_xlabel('Months of the years of the 21st century', fontsize = 15)
ax4.set_ylabel('D4 drought average, in %', fontsize = 15)

drought_data_US['year'] = drought_data_US.index.year
ax2.plot(drought_data_US.groupby(['year']).mean()['D4'])
ax2.set_title('Yearly D4 drought averages in the United States \n', fontsize = 15)
ax2.set_xlabel('Years of the 21st Century', fontsize = 15)
ax2.set_ylabel('D4 drought average, in %', fontsize = 15)

drought_data_US['month'] = drought_data_US.index.month 
ax5.plot(drought_data_US.groupby(['month']).mean()['D4'])
ax5.set_title('Monthly D4 drought averages in the United States \n', fontsize = 15)
ax5.set_xlabel('Months of the years of the 21st century', fontsize = 15)
ax5.set_ylabel('D4 drought average, in %', fontsize = 15)

drought_data_NV['year'] = drought_data_NV.index.year
ax3.plot(drought_data_NV.groupby(['year']).mean()['D4'])
ax3.set_title('Yearly D4 drought averages in Nevada \n', fontsize = 15)
ax3.set_xlabel('Years of the 21st Century', fontsize = 15)
ax3.set_ylabel('D4 drought average, in %', fontsize = 15)

drought_data_NV['month'] = drought_data_NV.index.month 
ax6.plot(drought_data_NV.groupby(['month']).mean()['D4'])
ax6.set_title('Monthly D4 drought averages in Nevada \n', fontsize = 15)
ax6.set_xlabel('Months of the years of the 21st century', fontsize = 15)
ax6.set_ylabel('D4 drought average, in %', fontsize = 15)

plt.show()
```

# Wildfire data analyzation

```python
wildfire = pd.read_csv('../FP04-groundwater/data/US_wildfires.csv')
for date in wildfire['Date']:
    date = str(date)
newdate = pd.to_datetime(wildfire['Date'], format = '%Y%m',errors = 'coerce')
wildfire['newdate'] = newdate
wildfire.index = wildfire['Date']
plt.figure(figsize = (10,10))
plt.plot(wildfire.index,wildfire['Acres Burned per Fire'])
x = np.array((wildfire.index))
lr = LinearRegression().fit(x.reshape(-1,1),wildfire["Acres Burned per Fire"])
plt.plot(x,lr.predict(x.reshape(-1,1)))
plt.show()
wildfire.index = newdate
wildfire['month'] = wildfire.index.month
wildfire['year'] = wildfire.index.year
month_plot = wildfire.groupby(['month']).mean()['Acres Burned']
year_plot = wildfire.groupby(['year']).mean()['Acres Burned']
plt.plot(month_plot)
plt.show()
plt.plot(year_plot)
del (wildfire['newdate'])
```
