```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gr_colo_spr = pd.read_csv("../data/groundwater_data/USGS_ColoradoSprings/WATERLEVEL.csv")
gr_colo_spr
```

```python
gr_colo_spr_sites = gr_colo_spr['SiteNo'].unique()
gr_colo_spr_sites
```

```python
# make dataframe for each site

gr_colo_spr_3856 = gr_colo_spr.loc[gr_colo_spr["SiteNo"] == 385607104230801]

gr_385344104242401 = gr_colo_spr.loc[gr_colo_spr["SiteNo"] == 385344104242401]

gr_384506104241302 = gr_colo_spr.loc[gr_colo_spr["SiteNo"] == 384506104241302]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(
    nrows = 3,
    ncols = 1,
    figsize = (15,10),
)

fig.suptitle("Groundwater Levels in Colorado Springs Area 2002-2023", fontsize = 30, y= 1)

ax1.plot(gr_colo_spr_3856["Time"], - gr_colo_spr_3856["Original Value"], linewidth = 5)
ax1.set_xticks(ticks = gr_colo_spr_3856["Time"][::12], labels = gr_colo_spr_3856["Time"][::12], rotation = 20)
ax1.set_title("Site 385607104230801")
#ax1.set_xlabel("Time")
ax1.set_ylabel("Water Level")

ax2.plot(gr_385344104242401["Time"], - gr_385344104242401["Original Value"], linewidth = 5)
ax2.set_xticks(gr_385344104242401["Time"][::12], gr_385344104242401["Time"][::12], rotation = 20)
ax2.set_title("Site 385344104242401")
#ax2.set_xlabel("Time")
ax2.set_ylabel("Water Level")

ax3.plot(gr_384506104241302["Time"], - gr_384506104241302['Original Value'], linewidth = 5)
ax3.set_xticks(gr_384506104241302['Time'][::12], gr_384506104241302["Time"][::12], rotation = 20)
ax3.set_title("Site 384506104241302")
ax3.set_ylabel("Water Level (Ft)")
ax3.set_xlabel("Time (DateTime")

fig.tight_layout()
fig.subplots_adjust(hspace=0.8)

plt.savefig("../Colorado_Springs_GWL_2002-2023")
```
