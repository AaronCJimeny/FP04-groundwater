```python
gr_colo_spr = pd.read_csv("../USGS_ColoradoSprings/WATERLEVEL.csv")

gr_colo_spr_filtered = gr_colo_spr[gr_colo_spr['Original Value'] < 2000]
gr_colo_spr_filtered.drop_duplicates()
gr_colo_spr_filtered['mean'] = gr_colo_spr_filtered.groupby("Time")['Original Value'].transform('mean')
gr_colo_spr_filtered = gr_colo_spr_filtered[gr_colo_spr_filtered['Original Value'].ne(gr_colo_spr_filtered['Original Value'].shift())]
gr_colo_spr_filtered
```

```python
gr_colo_spr_filtered = gr_colo_spr_filtered.sort_values(by='Time', ascending=True)
gr_colo_spr_filtered.index = pd.to_datetime(gr_colo_spr_filtered["Time"])
```
